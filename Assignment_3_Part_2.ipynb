```python
import pandas as pd
import matplotlib.pyplot as plt
ti = pd.read_csv ("titanic.csv")
print(ti.head())
```

```python
# a. Determine if the survival rate is associated to the class of passenger
# survival rate by class
survivors_class = ti.groupby('Pclass')['Survived'].mean()
print("Survival Rates by Passenger Class:")
print(survivors_class)
# plot
survivors_class.plot(kind='bar', rot=0)
plt.title('Survival Rate by Class of Passengers')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
# Findings: based on the analysis we can see that the higher the class of passenger, the higher chance of surviving.
```

```python
# b. Determine if the survival rate is associated to the gender
# survival rate by gender
survivors_sex = ti.groupby('Sex')['Survived'].mean()
print("Survival Rates by Gender:")
print(survivors_sex)
#plot
survivors_sex.plot(kind='bar',color=['pink','lightblue'], rot=0)
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()
# Findings: the survival rate of women is more than triple than that of men
```

```python
# c. Determine the survival rate is associated to the age
# survival rate by age
ti['Age'].fillna(ti['Age'].mean(), inplace=True) # clean data
# Grouping ages into categories
age_nums = [0, 12, 18, 30, 50, 100]  # Define age categories as needed
age_groups = ['0-12', '13-18', '19-30', '31-50', 'Over 50']
ti['AgeCategory'] = pd.cut(ti['Age'], bins=age_nums, labels=age_groups)
survivors_age = ti.groupby('AgeCategory')['Survived'].mean()
print("Survival Rates by Age:")
print(survivors_age)
# Plot
survivors_age.plot(kind='bar', color='purple', rot=0)
plt.title('Survival Rate by Age Category')
plt.xlabel('Age Category')
plt.ylabel('Survival Rate')
plt.show()
# Findings: Young children with the age of under 12 has the highest chance of survival. Followed by 13-18 and 31-50 age groups. Young adults between 19-30 years of age has the lowest chance of survial.
```
