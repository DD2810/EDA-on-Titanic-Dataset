```python
import pandas as pd
import matplotlib.pyplot as plt
tr = pd.read_csv ('train.csv')
print(tr.head())
```

```python
# a. Determine if there is a correlation between the price of houses increases with larger living area
cor = tr['GrLivArea'].corr(tr['SalePrice'])
print("Correlation between Living Area and House Price:", cor)
#plot
plt.scatter(tr['GrLivArea'], tr['SalePrice'], alpha=0.5, color = 'green')
plt.title('Living Area vs. House Price')
plt.xlabel('Living Area (sq. ft)')
plt.ylabel('House Price ($)')
plt.show()
```

```python
# b. Determine if there is a correlation between number of bedrooms with the house prices
cor_bed = tr['BedroomAbvGr'].corr(tr['SalePrice'])
print("Correlation between Bedrooms and House Price:", cor_bed)
#plot
plt.scatter(tr['BedroomAbvGr'], tr['SalePrice'], alpha=0.5, color = 'pink')
plt.title('Number of Bedrooms vs. House Prices')
plt.xlabel('Number of Bedrooms')
plt.ylabel('House Price ($)')
plt.show()
```

```python
# c.Will location (Neighborhood) affects house prices?
from sklearn.preprocessing import LabelEncoder # change data from catagorical to numerial
label_encoder = LabelEncoder()
tr['Neighborhood_Encoded'] = label_encoder.fit_transform(tr['Neighborhood'])
cor_neighborhood = tr['Neighborhood_Encoded'].corr(tr['SalePrice'])
print("Correlation between Neighborhood and House Price:", cor_neighborhood)

#plot
hood_price = tr.groupby('Neighborhood')['SalePrice'].mean().sort_values(ascending=False)
hood_price.plot(kind='bar', figsize=(10, 6))
plt.title('Average House Price by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Average House Price ($)')
plt.xticks(rotation=45)
plt.show()
```
